```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
foodb_nutr = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/FoodData_Central_csv_2022-10-28/food_nutrient.csv')
```

```python
# we only require the columns: fdc_id, nutrient_id and amount
foodb_nutr.drop(['data_points', 'derivation_id', 'min', 'max', 'median', 'loq', 'footnote', 'min_year_acquired'], axis=1, inplace = True)
foodb_nutr.info()
```

```python
# grouping all nutrients by unique fdc_id to make it readable

cols = ["fdc_id", "nutrient_id", "amount"]
temp = foodb_nutr[cols].groupby(cols[:-1]).agg(lambda x: x)
```

```python
temp.head()
```

```python
nutr_db = temp.unstack(level=1).droplevel(0, axis=1)
nutr_db.head()
```
